# Pre-Deal Checklist

## A. Market & Portfolio Readiness:
- Confirm market view & price context (JKM/TTF/Brent, spreads, forward curves).
- Verify portfolio availability & SPA obligations & flexibility.
- Confirm delivery terms (FOB vs DES).

## B. Shipping & Operations:
- Confirm vessel availability, suitability, and voyage constraints.
- Check loading/unloading windows and terminal slots.

## C. Economics & Risk:
- Validate deal margin including freight, BOG, port/terminal, storage/regas.
- Ensure hedging approach alignment.
- Complete regulatory/sanctions and destination clause checks.


## D. Credit & Counterparty:
- Check exposure vs limits in _Credit System_; validate need for collateral.
- Confirm credit filters alignment between `CTRM` and _Credit System_.

## E. Contracts & Terms:
- Review applicable SPA clauses and short-term agreement terms.
- Confirm payment, performance, and compliance items.


## F. Systems & Recordkeeping:
- Capture provisional deal in `CTRM` with correct entity & pricing.
- Confirm _Credit System_ deal status mapping.
- Secure trading, credit, shipping, and management approvals.

## G. Final Go/No-Go:
- Verify P&L, risk, credit, logistics, and approvals are all green.
